## Pandas DataFrame
image.png
**Series**

*   Uma variável
*   Informação de um único tipo
*   Preserva a consistência
*   Uma versão poderosa das listas do Python
*   Os índices podem ser relacionados à chave de um dicionário

**DataFrames**



*   Cada coluna representa uma variável diferente
*   Cada coluna tem seu tipo de dado diferente
*   A informação pode ser heterogênea
*   Preserva a consistência (Objetivo é ter um único tipo de variável em uma coluna específica)
*   Fornecemos um objeto inteiro que possui os valores de uma coluna inteira de chaves de dicionário
*   Herda as características da classe dos dicionários
*   Adiciona mais funcionalidades

### Criar um DataFrame a partir de um dicionário de Listas

```python
data = {
    'Nome': ['Ana', 'Carlos', 'Maria', 'Pedro'],
    'Idade': [23, 35, 29, 41],
    'Cidade': ['São Paulo', 'Rio de Janeiro', 'Curitiba', 'Porto Alegre']
}

# Criar DataFrame
df = pd.DataFrame(data)

print(df)
```

```python
data[0]['Nome']
```

### Criar um DataFrame a partir de uma lista de dicionários

```python
data = [
    {'Nome': 'Ana', 'Idade': 23, 'Cidade': 'São Paulo'},
    {'Nome': 'Carlos', 'Idade': 35, 'Cidade': 'Rio de Janeiro'},
    {'Nome': 'Maria', 'Idade': 29, 'Cidade': 'Curitiba'},
    {'Nome': 'Pedro', 'Idade': 41, 'Cidade': 'Porto Alegre'}
]

# Criar DataFrame
df = pd.DataFrame(data)

print(df)
```

### Criar DataFrame a partir de uma Lista de Listas

```python
# Lista de listas
data = [
    ['Ana', 23, 'São Paulo'],
    ['Carlos', 35, 'Rio de Janeiro'],
    ['Maria', 29, 'Curitiba'],
    ['Pedro', 41, 'Porto Alegre']
]

# Especificar os nomes das colunas
colunas = ['Nome', 'Idade', 'Cidade']

# Criar DataFrame
df = pd.DataFrame(data, columns=colunas)

print(df)
```

### Criar DataFrames a partir de arrays Numpy

```python
import numpy as np

# Arrays numpy
data = np.array([
    ['Ana', 23, 'São Paulo'],
    ['Carlos', 35, 'Rio de Janeiro'],
    ['Maria', 29, 'Curitiba'],
    ['Pedro', 41, 'Porto Alegre']
])

# Especificar os nomes das colunas
colunas = ['Nome', 'Idade', 'Cidade']

# Criar DataFrame
df = pd.DataFrame(data, columns=colunas)

print(df)
```

### Criar DataFrame Vazio e Adicionar Dados Posteriormente

```python
import pandas as pd

# Criar DataFrame vazio com colunas específicas
df = pd.DataFrame(columns=['Nome', 'Idade', 'Cidade'])

# Adicionar linhas ao DataFrame usando pd.concat()
nova_linha1 = pd.DataFrame([{'Nome': 'Ana', 'Idade': 23, 'Cidade': 'São Paulo'}])
nova_linha2 = pd.DataFrame([{'Nome': 'Carlos', 'Idade': 35, 'Cidade': 'Rio de Janeiro'}])
nova_linha3 = pd.DataFrame([{'Nome': 'Maria', 'Idade': 29, 'Cidade': 'Curitiba'}])
nova_linha4 = pd.DataFrame([{'Nome': 'Pedro', 'Idade': 41, 'Cidade': 'Porto Alegre'}])

# Concatenar as novas linhas ao DataFrame existente
df = pd.concat([df, nova_linha1], ignore_index=True)
df = pd.concat([df, nova_linha2], ignore_index=True)
df = pd.concat([df, nova_linha3], ignore_index=True)
df = pd.concat([df, nova_linha4], ignore_index=True)

print(df)
```

```python
# Buscar todas as linhas onde o nome é 'Ana'
resultado = df[df['Nome'] == 'Ana']

print("Busca na coluna 'Nome':")
print(resultado)
```
